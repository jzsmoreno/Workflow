# **Optimización**

* La palabra optimización hace referencia a la acción y efecto de optimizar. Podemos pensar a esta acción (**optimizar**) como aquel proceso que se encarga de resolver un problema de la forma más eficiente prosible y utilizando la menor cantidad de recursos.

Ejemplos de optimización:

1.   Encontrar la ruta mas corta para ir de león a acapulco
2.   Encontrar el mes del año en el cual la venta de autos maxima

* En matemáticas, ciencias de la computación, estadística, económia, etc. La palabra optimización se refiere a la acción de maximizar o minimizar una función (discreta o continua, n-dimensional) encontrando aquel valor en el dominio que hace minima o máxima a una función

1.   \begin{equation}

      \end{equation}

## **Método de Newton de primer orden**

* **El método de Newton es un método iterativo para encontrar aproximaciones de los ceros o raíces de una función real.**

Recordando la expansión en serie de Taylor de una función $f(x)$ alrededor de $x$.

\begin{equation}
      \begin{aligned}
      f(x + \delta x) \approx f(x) + (\delta x)\cdot\left(\frac{\mathrm{d} f(x)}{\mathrm{d} x}\right) \hspace{1em} (1)
      \end{aligned}
      \end{equation}

Entonces, el objetivo es encontrar $x^{*}$, tal que $f(x^{*}) = 0$

Definiendo: $\delta x = x_{k+1} - x_{k}$, donde $k$ es el paso de la iteración. Tenemos que:

\begin{equation}
      \begin{aligned}
      0 = f(x_{k}) + (x_{k+1} - x_{k})\cdot\left(\frac{\mathrm{d} f(x)}{\mathrm{d} x}\right) \hspace{1em} (2)
      \end{aligned}
      \end{equation}

despejando $x_{k+1}$

\begin{equation}
      \begin{aligned}
      x_{k+1} = x_{k} - \left(\frac{\mathrm{d} f(x)}{\mathrm{d} x}\right)^{-1}\cdot f(x_{k}) \hspace{1em} (3)
      \end{aligned}
      \end{equation}

## **Descenso del Gradiente 2-dimensional**

* **El descenso del gradiente es un método iterativo de primer orden cuyo objetivo es encontrar minimos locales de una función diferenciable.** 


Derivada : Es la razón del incremento de la variable dependiente $\left(f(x + h) - f(x)\right)$ entre el incremento de la variable independiente $\left((x+h) - x\right)$, donde vamos a exigir que el incremento sea muy pequeño.

\begin{equation}
      \begin{aligned}
      f'(x) = \lim\limits_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h} \hspace{1em} (4)\\
      \end{aligned}
      \end{equation}

La derivada nos da información del cambio de un función, nos dice cuando la función esta creciendo o decreciendo.

Los puntos críticos de una función $f(x)$ se calculan resolviendo $\frac{\mathrm{d} f(x)}{\mathrm{d} x} = 0$. Los puntos críticos pueden ser minimos o máximos locales o globales, o puntos de inflexión.

La derivada $f' (x)$ es una función diferente a $f(x)$. Por ejemplo: $f(x) = x^{2}$ y $f'(x) = 2x$ son funciones distintas. 

Entonces, utilizando la ecuación 1 y 3, podemos encontrar iterativamente puntos críticos $x^{*}$ de $f(x)$

\begin{equation}
      \begin{aligned}
      f' (x + \delta x) \approx f' (x) + (\delta x)\cdot\left(\frac{\mathrm{d}^{2} f(x)}{\mathrm{d} x^{2}}\right) \hspace{1em} (5)
      \end{aligned}
      \end{equation}

El objetivo es encontrar $x^{*}$, tal que $f '(x^{*}) = 0$

Definiendo: $\delta x = x_{k+1} - x_{k}$, donde $k$ es el paso de la iteración. Tenemos que:

\begin{equation}
      \begin{aligned}
      0 = f' (x_{k}) + (x_{k+1} - x_{k})\cdot\left(\frac{\mathrm{d}^{2} f(x)}{\mathrm{d} x^{2}}\right) \hspace{1em} (6)
      \end{aligned}
      \end{equation}

despejando $x_{k+1}$

\begin{equation}
      \begin{aligned}
      x_{k+1} = x_{k} - \left(\frac{\mathrm{d}^{2} f(x)}{\mathrm{d} x^{2}}\right)^{-1}\cdot f' (x_{k}) \hspace{1em} (7)
      \end{aligned}
      \end{equation}

### **Ejemplo: Encontrar un punto crítico de la siguiente función**
\begin{equation}
      \begin{aligned}
      f(x) = x^{2} - 2x \hspace{1em} (8)
      \end{aligned}
\end{equation}

1.   Analíticamente:
\begin{equation}
      \begin{aligned}
      \frac{\mathrm{d} f(x)}{\mathrm{d} x} = 2x - 2; \hspace{0.5em} donde \hspace{0.5em} \frac{\mathrm{d} f(x)}{\mathrm{d} x} = 0.
      \end{aligned}
\end{equation}

Entonces $x^{*} = 1$ es un punto crítico de $f(x)$. Que en este caso es un minimo global de la función $f(x)$

2.   Iterativamente:

\begin{equation}
      \begin{aligned}
      x_{k+1} = x_{k} - \left(\frac{1}{2}\right)\cdot (2(x_{k}) - 2) \hspace{1em} (8)
      \end{aligned}
\end{equation}

Inicializando en $x_{0} = 0.5$

\begin{equation}
      \begin{aligned}
      x_{1} = 0.5 - \left(\frac{1}{2}\right)\cdot (-1) \hspace{1em} (9)
      \end{aligned}
      \end{equation}

Encontramos $x^{*} = x_{1}$

De tal forma, definimos el descenso del gradiente 2-dimensional:

\begin{equation}
      \begin{aligned}
      x_{k+1} = x_{k} - \alpha\cdot f' (x_{k}) \hspace{1em} (10)
      \end{aligned}
      \end{equation}

donde $\alpha = \left(\frac{\mathrm{d}^{2} f(x)}{\mathrm{d} x^{2}}\right)^{-1}$ es una constante. En manchine learning se le conoce como tasa de aprendizaje. 

## **Descenso del Gradiente n-dimensional**

* El descenso del gradiente 2-dimensional minimiza funciones de la forma $f : \mathbb{R} \mapsto \mathbb{R}$, por ejemplo $f(x) = x^{2}$.

* Consideramos ahora funciones más generales de la forma $F : \mathbb{R}^{n} \mapsto \mathbb{R}$, por ejemplo $F(x,y) = x^{2} + y^{2}$


La derivada en funciones $n-dimensionales$ se llama vector gradiente (ecuación 11)

\begin{equation}
      \begin{aligned}
      \nabla F(X) = \begin{bmatrix}
                  \frac{\partial F }{\partial x_{1}}&..&\frac{\partial F}{\partial x_{n}} 
                  \end{bmatrix} \hspace{1em} (11)
      \end{aligned}
      \end{equation}


Extrapolando las ecuaciones $1,5,7,10$ a funciones $n-dimensionales$, tenemos que:

\begin{equation}
      \begin{aligned}
      F(X + \Delta X) \approx F(X) + (\Delta X)^{T}\nabla F(X) \hspace{1em} (12)
      \end{aligned}
      \end{equation}


\begin{equation}
      \begin{aligned}
      \nabla F(X + \Delta X) \approx  \nabla F(X) + (\Delta X)^{T}H(x)(\Delta X) \hspace{1em} (13)
      \end{aligned}
      \end{equation}


\begin{equation}
      \begin{aligned}
      X_{k+1} = X_{k} - H^{-1}(X_{k})\nabla F(X_{k}) \hspace{1em} (14)
      \end{aligned}
      \end{equation}

Entonces, definimos el descenso del gradiente $n-dimensional$:

\begin{equation}
      \begin{aligned}
      X_{k+1} = X_{k} - \alpha\cdot \nabla F(X_{k}) \hspace{1em} (15)
      \end{aligned}
      \end{equation}


Donde $X = \begin{bmatrix} x_{1}&..&x_{n}\end{bmatrix}$

### **Aplicación del descenso del gradiente para regresión lineal 2-dimensional**

* El objetivo de la regresión lineal 2-dimensional es encontrar los parámetros $a,b$ del modelo $f(x;a,b) = a\cdot x + b$

* Consideramos el siguiente conjunto de entrenamiento:

\begin{equation}
      \begin{aligned}
      D = \bigl(\begin{smallmatrix} (x_{1},y_{1}),&...&,(x_{m},y_{m}) \end{smallmatrix}\bigr) \hspace{1em} (16)
      \end{aligned}
      \end{equation}

* Definimos la siguiente funcion de costo ($J : \mathbb{R}^{2} \mapsto \mathbb{R}$):

\begin{equation}
      \begin{aligned}
      J(a,b) = \frac{1}{2m}\sum_{i=1}^{n}(f(x_{i};a,b) - y_{i})^2 \hspace{1em} (17)
      \end{aligned}
      \end{equation}

* Definimos el siguiente problema de minimización:

\begin{equation}
      \begin{aligned}
      \min_{a,b} \quad & J(a,b) \hspace{1em} (18)
      \end{aligned}
      \end{equation}


* Computamos el gradiente $\nabla J(a,b)$:


\begin{equation}
      \begin{aligned}
      \nabla J(a,b) = \begin{bmatrix}
                  \frac{\partial J}{\partial a},\frac{\partial J}{\partial b} 
                  \end{bmatrix} \hspace{1em} (19) 
      \end{aligned}
      \end{equation}


\begin{equation}
      \begin{aligned}
      \nabla J(a,b) = \begin{bmatrix}
                  \frac{1}{m}\sum_{i=1}^{n}(f(x_{i};a,b) - y_{i}), \frac{1}{m}\sum_{i=1}^{n}(f(x_{i};a,b) - y_{i})\cdot x_{i} 
                  \end{bmatrix} \hspace{1em} (20)
      \end{aligned}
      \end{equation}


Entonces, haciendo uso del descenso del gradiente (ecuación 21), podemos encontrar los parámetros $a^*,b^*$ que minimicen la función de costo $J(a,b)$

Sea $X = \begin{bmatrix} a,b\end{bmatrix}$, tenemos que:

\begin{equation}
      \begin{aligned}
      X_{k+1} = X_{k} - \alpha\cdot \nabla J(X_{k}) \hspace{1em} (21)
      \end{aligned}
      \end{equation}

Podemos ver a la ecuación 21, como un sistema de ecuaciones, de la siguiente forma:

\begin{equation}
      \begin{aligned}
      a_{k+1} = a_{k} - \alpha\cdot \frac{1}{m}\sum_{i=1}^{n}(f(x_{i};a_{k},b_{k}) - y_{i}) \hspace{1em} (22)
      \end{aligned}
      \end{equation}
\begin{equation}
      \begin{aligned}
      b_{k+1} = b_{k} - \alpha\cdot \frac{1}{m}\sum_{i=1}^{n}(f(x_{i};a_{k},b_{k}) - y_{i})\cdot x_{i}) \hspace{1em} (23)
      \end{aligned}
      \end{equation}



```python
X = np.linspace(-20.0, 20.0, 200)
Y = np.copy(X)
theta = [2.0, 5.0]
x_vec = np.array([X, Y])


def model(x, theta):
    return theta[0] * (x[0] - 1.0) ** 2.0 + theta[1] * (x[1] - 1.0) ** 2.0
```

```python
X = np.linspace(-100.0, 100.0, 200)
x, y = np.meshgrid(X, X)


def f(x, y):
    return 2.0 * (x - 1) ** 2.0 + 5.0 * (y - 1) ** 2.0


z = f(x, y)

ax = plt.axes(projection="3d")
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap="viridis", edgecolor="none")
ax.view_init(25, 45)
ax.set_title("$f(x, y) = 2(x-1)^{2}+5(y-1)^{2}$")
# ax.set_ylabel(r'$x_{2}$')
# ax.set_xlabel(r'$x_{1}$')
# ax.set_zlabel('z')
plt.show()
```

```python
def cal_cost(model, theta, x, om):
    val_min = model(x, theta)
    y = model(om, theta)

    cost = np.sqrt(np.square(val_min - y))

    return cost


def grad_descent(model, theta, vec_int, rate, inter, sigma, tol, n=10):
    om = np.array([(2.0 * np.random.rand() - 1) * n, (2.0 * np.random.rand() - 1) * n])
    cost_hist = []
    min_hist = []
    for i in range(inter):
        print("Comienza metodo :", i)
        print("vec_int:", vec_int[0], vec_int[1])
        min_hist.append(vec_int)
        cost = cal_cost(model, theta, vec_int, om)
        cost_hist.append(cost)
        x2 = [om[0], vec_int[1]]
        y2 = [vec_int[0], om[1]]
        x1 = [vec_int[0], vec_int[1]]
        y1 = [vec_int[0], vec_int[1]]
        delta = np.array([x1[0] - x2[0], y1[1] - y2[1]])
        print("delta:", delta)
        gradx_1 = model(x1, theta) - model(x2, theta)
        gradx_2 = model(y1, theta) - model(y2, theta)
        grad = np.array([gradx_1, gradx_2])
        print("grad:", grad)
        if 0.0 < np.sqrt(gradx_1**2.0 + gradx_2**2.0) < tol and model(om, theta) < model(
            vec_int, theta
        ):
            print("Goal reached :", om)
            return (om, cost_hist, np.array(min_hist))
        else:
            if model(om, theta) < model(vec_int, theta):
                vec_int = vec_int - rate * delta
            else:
                x_1 = np.random.normal(vec_int[0], sigma, 1)[0]
                x_2 = np.random.normal(vec_int[1], sigma, 1)[0]
                om = np.array([x_1, x_2])
                if model(om, theta) < model(vec_int, theta):
                    x1 = [om[0], vec_int[1]]
                    y1 = [vec_int[0], om[1]]
                    gradx_1 = model(x2, theta) - model(x1, theta)
                    gradx_2 = model(y2, theta) - model(y1, theta)
                    grad = np.array([gradx_1, gradx_2]) * (model(om, theta) / model(vec_int, theta))
                    delta = np.array([x2[0] - x1[0], y2[1] - y1[1]])
                    vec_int = vec_int - rate * delta
                else:
                    ratio = model(vec_int, theta) / model(om, theta)
                    if ratio < np.random.rand():
                        x_1 = np.random.normal(vec_int[0], sigma, 1)[0]
                        x_2 = np.random.normal(vec_int[1], sigma, 1)[0]
                        om = np.array([x_1, x_2])
                    else:
                        vec_int = vec_int - rate * delta
        clear_output(wait=True)
    print("Máx. number of iterations reached :", om)
    return (om, cost_hist, np.array(min_hist))
```

```python
def SGD(model, theta, vec_int, inter, sigma, tol):
    om = np.array(
        [np.random.normal(vec_int[0], sigma, 1)[0], np.random.normal(vec_int[1], sigma, 1)[0]]
    )
    cost_hist = []
    min_hist = []
    for i in range(inter):
        print("Comienza metodo :", i)
        print("vec_int:", vec_int[0], vec_int[1])
        min_hist.append(vec_int)
        cost = cal_cost(model, theta, vec_int, om)
        cost_hist.append(cost)
        x2 = [om[0], vec_int[1]]
        y2 = [vec_int[0], om[1]]
        x1 = [vec_int[0], vec_int[1]]
        y1 = [vec_int[0], vec_int[1]]
        gradx_1 = model(x1, theta) - model(x2, theta)
        gradx_2 = model(y1, theta) - model(y2, theta)
        grad = np.array([gradx_1, gradx_2])
        mod = np.sqrt(grad[0] ** 2.0 + grad[1] ** 2)
        print("mod:", mod)
        if mod <= tol:
            print("Goal reached :", om)
            return (om, cost_hist, np.array(min_hist))
        else:
            if model(om, theta) < model(vec_int, theta):
                vec_int = om
                om = np.array(
                    [
                        np.random.normal(vec_int[0], sigma, 1)[0],
                        np.random.normal(vec_int[1], sigma, 1)[0],
                    ]
                )
            else:
                om = np.array(
                    [
                        np.random.normal(vec_int[0], sigma, 1)[0],
                        np.random.normal(vec_int[1], sigma, 1)[0],
                    ]
                )

        clear_output(wait=True)
    print("Máx. number of iterations reached :", om)
    return (om, cost_hist, np.array(min_hist))
```
