## **Máquinas de vector soporte**

* Las máquinas de vectores de soporte o máquinas de vector soporte (del inglés Support Vector Machines, SVM) son un conjunto de algoritmos de aprendizaje supervisado desarrollados por el matemático Vladimir Vapnik en la decada de los 90.

* Las máquinas de vector soporte estan pensadas para resolver problemas de regresion y clasificación.

* En esta sesión explicaremos intuitivamente que son las máquinas de soporte vectorial y su aplicación en problemas de regresión.

### **Máquinas de Vector Soporte para regresión**










* Consideramos un conjunto de ejemplos de entrenamientos $X = ${$(x_{1},y_{1}), .....,(x_{n},y_{n}) $}$ $, donde $x_{i} \in \mathbb R^d$ e $y_{i} \in \mathbb R$.

* El objetivo de las máquinas de soporte vectorial es encontrar los parámetros $(w_{1},....,w_{d},b)$ que mejor se ajusten a los datos de acuerdo con el siguiente modelo lineal:

\begin{equation} 
    f(x) = w_{1}x_{1}+w_{2}x_{2}+...+w_{d}x_{n} + b
\end{equation}

* Si el espacio de la variable independiente es $d=1$, entonces trataríamos de encontrar la siguiente recta: $y = w_{1}x_{1} + b$. Si el espacio fuera $d = 2$ para la variable independiente trataríamos de encontrar el siguiente plano: $y = w_{1}x_{1} + w_{2}x_{2} + b$. Si el espacio es $d>2$ entonces trataríamos de encontrar hiperplanos.


## **Máquinas vector soporte para datos con dependencia lineal**

* Supongamos que tenemos los siguientes datos de entrenamiento: $X={(x_{1},y_{1}),.....,(x_{n},y_{n})}$, donde $x_{i} \in \mathbb R$ e $y_{i} \in \mathbb R$

Supóngase entonces que se tiene un vector $\vec{w}$ el cual es siempre perpendicular al hiperplano que se desea encontrar. De manera que podemos determinar que tan lejos esta dicho vector de cada uno de nuestros datos.

\begin{equation}
\frac{\vec{d_{j}}\cdot \vec{w}}{\left | w \right |}=\left | d_{j} \right |\cos\theta
\end{equation}

donde $\vec{d_{j}}$ es el vector de distancia del punto $j$ al hiperplano con normal $\vec{w}$. En este sentido la proyección de cualquier vector $\vec{x}$ cumple que $\left \langle w, x \right \rangle-b = 0$ para todo punto $x_{j}$ en el plano. (donde $\left \langle w, x \right \rangle = \vec{w}\cdot \vec{x}$ es llamado el kernel del sistema)

*   $y_{i} = -\epsilon$, para $\left ( \vec{w}\cdot \vec{x}-b\leq  0 \right )$
*   $y_{i} = +\epsilon$, para $\left ( \vec{w}\cdot \vec{x}-b\geq  0 \right )$

o también,

\begin{equation}
y_{i}\left ( \vec{w}\cdot \vec{x}-b \right )\geq  \epsilon
\end{equation}

de la ecuación anterior tenemos dos condiciones


1.   $y_{i}\left ( \vec{w}\cdot \vec{x_{+}}-b \right )\geq  \epsilon$
2.   $y_{i}\left ( \vec{w}\cdot \vec{x_{-}}-b \right )\geq  \epsilon$

restando los puntos que se encuentran una distancia $+\epsilon$ con los que se encuentran a una distancia $-\epsilon$ por encima del hiperplano, dará por resultado dos veces el ancho de la distancia entre margen y margen. Así pues para los puntos que se encuentran sobre éstos márgenes, podemos describirlos matemáticamente como le sigue, siendo los puntos sobre el margen inferior:

\begin{equation}
\left ( \vec{w}\cdot \vec{x_{+}}-b \right )-\epsilon = 0
\end{equation}

Análogamente para el margen superior:

\begin{equation}
\left ( \vec{w}\cdot \vec{x_{-}}-b \right )+\epsilon = 0
\end{equation}

entonces de estas dos ecuaciones podemos deducir lo siguiente.

\begin{equation}
\vec{w}\cdot \vec{x_{+}} = b+\epsilon
\end{equation}

\begin{equation}
\vec{w}\cdot \vec{x_{-}} = b-\epsilon 
\end{equation}

donde ahora si somos capaces de escribir una ecuación que nos determine el ancho

\begin{equation}
\frac{\left ( b+\epsilon -\left [ b-\epsilon  \right ] \right )}{\left | \vec{w} \right |} = \left ( \vec{x_{+}}-\vec{x_{-}} \right )\cdot \frac{\vec{w}}{\left | \vec{w} \right |}
\end{equation}

\begin{equation}
\therefore ancho = \frac{ 2\epsilon}{\left | \vec{w} \right |}
\end{equation}

y dado que el objetivo de una máquina de soporte vectorial, es el de maximizar la anchura de dichos márgenes, dicho problema es equivalente al de minimizar $\left | \vec{w} \right |$, por lo tanto se puede escribir de la siguiente manera:

\begin{equation}
\min f(w) = \frac{1}{2}\left | \vec{w} \right |^{2}
\end{equation}

definimos entonces una función que incluye las restricciones del problema a minimizar la cual llamaremos $L$ y definiremos como:

\begin{equation}
L = \frac{1}{2}\left | \vec{w} \right |^{2}-\sum_{i}^{n}\alpha _{i}\left [ y_{i}\left ( \vec{w}\cdot \vec{x}-b \right )-\epsilon  \right ]
\end{equation}

y veremos que al derivar con respecto a la variable de interés $w$ exigiendo que sea un valor extremal igualando a cero dicha derivada.

\begin{equation}
\frac{\partial L}{\partial w} = \vec{w}-\sum_{i}^{n}\alpha _{i}y_{i}x_{i} = 0
\end{equation}

entonces despejando para $\vec{w}$

\begin{equation}
\vec{w}=\sum_{i}^{n}\alpha _{i}y_{i}x_{i}
\end{equation}

Tomando la segunda restricción se obtiene lo siguiente:

\begin{equation}
\frac{\partial L}{\partial b} = \sum_{i}^{n}\alpha _{i}y_{i} = 0
\end{equation}

Al sustituir la expresión para $\vec{w}$ en $L$ 

\begin{equation}
L = \frac{1}{2}\left ( \sum_{i}^{n}\alpha _{i}y_{i}x_{i} \right )\cdot \left ( \sum_{i}^{n}\alpha _{i}y_{i}x_{i} \right )-\sum_{i}^{n}\sum_{j}^{n}\alpha _{i}y_{i}\alpha _{j}y_{j}\left ( x_{i}x_{j} \right )-\sum_{i}^{n}\alpha _{i}y_{i}b+\sum_{i}^{n}\alpha _{i}
\end{equation}


Simplificando términos

\begin{equation}
L = \sum_{i}^{n}\alpha _{i}-\frac{1}{2}\sum_{i}^{n}\sum_{j}^{n}\alpha _{i}y_{i}\alpha _{j}y_{j}\left ( x_{i}x_{j} \right )
\end{equation}

Escribiéndolo en forma compacta

\begin{equation}
L = \alpha -\frac{1}{2}\left ( \alpha ^{T}X \right )\cdot \left ( \alpha ^{T}X \right ) = \alpha -\frac{1}{2}\left ( \alpha ^{T}X \right )^{T}\cdot \left ( X\alpha ^{T} \right ) = \alpha -\frac{1}{2}\left ( \alpha X^{T} \right )\cdot \left ( X\alpha ^{T} \right )
\end{equation}

Definiendo el tensor $P = X^{T}X$ 

\begin{equation}
L = \alpha-\frac{1}{2}\alpha P\alpha ^{T}
\end{equation}

donde $X = \left [ x_{1}y_{1}, x_{2}y_{2}, ..., x_{n}y_{n} \right ]$ 

```
# Importamos las librerías a utilizar
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```
plt.rcParams['font.size'] = 24
plt.style.use('seaborn-poster')
```

```
def f(x, epsilon):  # Función f(x) = 0.1*x + 1.25 + epsilon*ruido_gaussiano
    np.random.seed(1) # Definimos la semilla
    y = 0.1*x + 1.25 + epsilon*np.random.randn(x.shape[0])
    return y

def getLines(n, x, label = False):
  np.random.seed(1)
  for ni in range(n):
    m = 0.1 + np.random.normal(0, 0.025)
    b = 1.25 + np.random.normal(0, 0.025)
    y =  m*x + b

    if label:
      plt.plot(x,y, '--', label = "f(x) = {0:.2f}x + {1:.2f}".format(m, b))
      plt.legend()
    else:
      plt.plot(x,y)
```

```
""" Generamos valores de la variable dependiente x """

x = np.arange(0, 20, 0.5) # Generamos valores x de 0 a 20 en intervalos de 0.5 -> [0,0.5,.....,15]
y = f(x, 0.2) # Calculamos y, a partir de la función que hemos generado

# Hacemos un gráfico de los datos que hemos generado
plt.scatter(x,y, label='data', color='blue')
plt.xlabel("$X_{1}$")
plt.ylabel("$y$")
plt.title('Gráfica 1')
plt.legend()
plt.show()
```

* Nos preguntamos, ¿Cuál es la mejor función de la forma $f(x) = w_{1}x + b$  que se ajusta a los datos de la Figura 2?

```
plt.scatter(x,y, label='data', color='blue')
plt.xlabel("$X_{1}$")
plt.ylabel("$y$")
plt.title('Gráfica 2')

n_functions = 5 # Número de ajustes
getLines(n_functions,x, label = True)
plt.show() 
```

Bosquejemos la siguiente idea: 

* Definamos el margen $\epsilon$ de la función $f(x) = w_{1}x + b$ como la distancia ortogonal en ambos lados de la función $f(x)$ a un punto $(x_{j},y_{j})$. De modo que se forman dos rectas paralelas a la función $f(x)$ (ver Figura 3)

* Los ejemplos de entrenamieno $(x_{i},y_{i})$ que yacen fuera de las rectas del margen se les llama vectores de soporte


```
plt.scatter(x,y, label='data', color='blue')
plt.xlabel("$X_{1}$")
plt.ylabel("$y$")
plt.title('Gráfica 3')

np.random.seed(1)
m = 0.1 + np.random.normal(0, 0.005)
b = 1.25 + np.random.normal(0, 0.005)
f =  m*x + b

epsilon = 0.2
f_1 =  m*x + b + epsilon
f_2 =  m*x + b - epsilon

plt.plot(x,f, label = "f(x) = $w_1$*x + b", color = "black")
plt.plot(x,f_1, '--', label = "f(x) = $w_1$*x + b + epsilon", color = "red")
plt.plot(x,f_2, '--', label = "f(x) = $w_1$*x + b - epsilon", color = "red")
plt.legend()
plt.show()
```

Las máquinas de soporte vectorial tiene dos objetivos: 
  1.   Encontrar los parámetros $(w_{1},..,w_{d},b)$
  2.   Encontrar un margen $\epsilon$ lo suficientemente pequeño que permita abarcar o envolver la mayor cantidad de ejemplos de entrenamiento dentro del margen. Para que a su vez el encontrar los parámetros $(w_{1},..,w_{d},b)$ del modelo es equivalente a encontrar un margen $\epsilon$  

## **Máquinas vector soporte para datos con dependencia no lineal**

Supongamos que tenemos los siguientes datos de entrenamiento: $X={(x_{1},y_{1}),.....,(x_{n},y_{n})}$, donde $x_{i} \in \mathbb R$ e $y_{i} \in \mathbb R$

Existen diferentes tipos de kernels de entre los cuales analizaremos el **polinomial** y el **RBF** o función radial básica.



*   Para kernels polinomiales $K\left ( w, x \right ) = \left ( \left \langle w, x \right \rangle-b \right )^{d}$ donde $d$ es el grado del polinomio.
*   El kernel de la función radial básica $K\left ( w, x \right ) = \exp\left ( {-\frac{1}{2\sigma ^{2}}\left \| w-x \right \|^{2}} \right )$


```
# Generamos los siguientes datos
np.random.seed(1)
m = 80
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**3 + 0.1*X**2 + 0.3*X + 2 + np.random.randn(m, 1)

plt.scatter(X,y, label='data', color='blue')
plt.xlabel("$X_{1}$")
plt.ylabel("$y$")
plt.title('Gráfica 4')
plt.legend()
plt.show()
```

Como en el caso lineal, el objetivo es construir un modelo que tenga la característica de minimizar el error o diferencia, para el caso de una regresión, dicho error, se cuantifica por medio del error cuadrático medio de los valores estimados por el modelo y los valores reales.

```
""" Construimos los polinomio que ajustan a los datos de la gráfica 4 """

x = np.arange(-3, 3, 0.1)
f = 0.5 * x**3 + 0.1*x**2 + 0.3*x + 2 
epsilon = 1
f_1 = 0.5 * x**3 + 0.1*x**2 + 0.3*x + 2 + epsilon
f_2 = 0.5 * x**3 + 0.1*x**2 + 0.3*x + 2 - epsilon

plt.scatter(X,y, label='data', color='blue')
plt.xlabel("$X_{1}$")
plt.ylabel("$y$")
plt.title('Gráfica 5')

plt.plot(x,f, label = "Model", color = "black")
plt.plot(x,f_1, '--', label = "Model + epsilon", color = "red")
plt.plot(x,f_2, '--', label = "Model - epsilon", color = "red")
plt.legend()
plt.show()
```

* Como hemos estudiado, cuando los datos tienen una relación de dependencia no lineal, el número de parámetros a encontrar es mayor (Para el caso de la regresión polinomial por ejemplo), y por consecuencia el modelo es más complejo.

* Sorprendentemente, las máquinas vector soporte logran mantener el número de parámetros a encontrar independientemente de la relación lineal o no, de los datos. 

* Por poner un ejemplo si nos basamos en la Gráfica 5, y quisiéramos encontrar un modelo polinomial que ajustar a esos datos, tendríamos que encontrar posiblemente entre 2 a 4 parámetros (grados de polinomio). Sin embargo, en las máquinas de soporte vectorial no necesitamos aumentar la complejidad del modelo, al igual que en el caso lineal (Gráfica 3) solo necesitamos encontrar 2 parámetros para ajustar los datos de la Gráfica 5.

* Consideremos la siguiente idea (Gráfica 6, 7, 8 y 9)

```
# Transformación de los datos
fig, axs = plt.subplots(2, 2, figsize = (20, 16))

axs[0, 0].scatter(X,y, label='Original data', color='blue')
axs[0, 0].set_title('Gráfica 6')
axs[0, 0].set(xlabel='$X_{1}$', ylabel='$y_{1}$')
axs[0, 0].ticklabel_format(style='plain')

x = np.arange(0, 20, 0.2)
f_tr =  0.1*x + 1.25 + 0.1*np.random.randn(x.shape[0])
axs[0, 1].scatter(x,f_tr, label = "Transformed data", color = "blue", marker = "v")
axs[0, 1].set_title('Gráfica 7')
axs[0, 1].set(xlabel='$X_{2}$', ylabel='$y_{2}$')
axs[0, 1].ticklabel_format(style='plain')

x1 = np.arange(-3, 3, 0.1)
f = 0.5 * x1**3 + 0.1*x1**2 + 0.3*x1 + 2 
epsilon = 1
f_1 = 0.5 * x1**3 + 0.1*x1**2 + 0.3*x1 + 2 + epsilon
f_2 = 0.5 * x1**3 + 0.1*x1**2 + 0.3*x1 + 2 - epsilon

axs[1, 0].plot(x1,f, label = "Model original data", color = "black")
axs[1, 0].scatter(X,y, color='blue')
axs[1, 0].plot(x1,f_1, '--', label = "Model original data + epsilon", color = "red")
axs[1, 0].plot(x1,f_2, '--', label = "Model original data - epsilon", color = "red")
axs[1, 0].set_title('Gráfica 8')
axs[1, 0].set(xlabel='$X_{1}$', ylabel='$y_{1}$')
axs[1, 0].ticklabel_format(style='plain')

epsilon = 0.1
ff =  0.1*x + 1.25
ff_1 = 0.1*x + 1.25 + epsilon
ff_2 = 0.1*x + 1.25 - epsilon

axs[1, 1].plot(x,ff, label = "Model Transformed data", color = "green")
axs[1, 1].scatter(x,f_tr, color='blue', marker = "v")
axs[1, 1].plot(x,ff_1, '--', label = "Model Transformed data + epsilon", color = "gray")
axs[1, 1].plot(x,ff_2, '--', label = "Model Transformed data - epsilon", color = "gray")
axs[1, 1].set_title('Gráfica 9')
axs[1, 1].set(xlabel='$X_{2}$', ylabel='$y_{2}$')
axs[1, 1].ticklabel_format(style='plain')

fig.legend()
fig.show()
```

Entonces, para datos con dependencia no lineal (Figura 6), las máquinas de soporte que transforman los datos a un espacio donde los nuevos datos transformados son lineales (Figura 7). En el nuevo espacio $(X_{2}, y_{2})$ de la Figura 7, se encuentran los parámetros del modelo lineal, como se hizó en la Figura 5; una vez encontrado el modelo lineal se puede hacer un mapeo $(X_{1},y_{1}) <-> (X_{2}, y_{2})$

### **Ejemplo**

Supóngase que se cuenta con los datos de la sesión anterior, donde claramente tenemos una dependencia no lineal. Veamos como implementar las máquinas de soporte vectorial.

```
np.random.seed(1)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()
```

```
def plot_svm_regression(svm_reg, X, y, axes):
    x1s = np.linspace(axes[0], axes[1], 100).reshape(100, 1)
    y_pred = svm_reg.predict(x1s)
    plt.plot(x1s, y_pred, "r-", linewidth=2, label=r"$\hat{y}$")
    plt.plot(x1s, y_pred + svm_reg.epsilon, "k--")
    plt.plot(x1s, y_pred - svm_reg.epsilon, "k--")
    plt.scatter(X[svm_reg.support_], y[svm_reg.support_], s=280, facecolors='#FFAAAA')
    plt.plot(X, y, "bo")
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.legend(loc="upper left", fontsize=18)
    plt.axis(axes)
```

```
value = 0.1

svm_poly_reg1 = SVR(kernel="poly", degree=2, C=100, epsilon=value)
svm_poly_reg1.fit(X, y)

svm_poly_reg2 = SVR(kernel="poly", degree=2, C=0.01, epsilon=value)
svm_poly_reg2.fit(X, y)

fig, axes = plt.subplots(ncols=2, figsize=(18, 6), sharey=True)
plt.sca(axes[0])
plot_svm_regression(svm_poly_reg1, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg1.degree, svm_poly_reg1.C, svm_poly_reg1.epsilon))
plt.xlabel('$X_{1}$')
plt.ylabel('$y$')
plt.sca(axes[1])
plot_svm_regression(svm_poly_reg2, X, y, [-1, 1, 0, 1])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg2.degree, svm_poly_reg2.C, svm_poly_reg2.epsilon))
plt.legend()
plt.show()
```

### **Ejercicio** 

Ahora discutamos un caso particular en el que la transformación a realizar no es simplemente la de un polinomio, sino que utilizaremos por kernel una función llamada **función Radial Básica**, la cual nos permitirá extender las máquinas de soporte vectorial a datos que posean una dependencia armónica.

```
# CREAR EL DATASET
m = 200
X_test = np.linspace(-3, 3, 200)
X_test = np.array(X_test)
X_test = X_test[np.newaxis,...].T
np.random.seed(1)
y_train = (0.2 + np.cos(0.2*X_test) + 0.5 * np.sin(4.0*X_test) + np.random.randn(m, 1)/5 + 0.5*X_test).ravel()
plt.plot(X_test, y_train, 'bo', label = 'dataset')
value = 0.2
svm_poly_reg3 = SVR(kernel="rbf", gamma=2, epsilon=value)
svm_poly_reg3.fit(X_test, y_train)

plot_svm_regression(svm_poly_reg3, X_test, y_train, [min(X_test), max(X_test), min(y_train), max(y_train)])
plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg3.degree, svm_poly_reg3.C, svm_poly_reg3.epsilon))

plt.ylabel('y')
plt.xlabel('$X_{1}$')
plt.legend()
plt.show()

#svm_poly_reg4 = SVR(kernel="rbf", C=50, gamma = 10, epsilon=value)
#svm_poly_reg4.fit(X, y)

#plot_svm_regression(svm_poly_reg4, X, y, [min(X), max(X), min(y), max(y)])
#plt.title(r"$degree={}, C={}, \epsilon = {}$".format(svm_poly_reg4.degree, svm_poly_reg4.C, svm_poly_reg4.epsilon))
```

La prueba de hipotésis utilizada para cuantificar el error de una regresión, que comúnmente se emplea, es la prueba **Chi-cuadrada** cuyo estadístico se define como le sigue:  

$\chi ^{2} = \sum_{i}^{n}\frac{\left ( y_{i}-f(x_{i}) \right )^{2}}{\Delta x_{i}^{2}+\Delta y_{i}^{2}}$

donde $0\leq \chi ^{2}\leq 1$ para $1$, el mejor ajuste y $0$ el peor ajuste.

```
svm_poly_reg3.score(X_test, y_train)
```
